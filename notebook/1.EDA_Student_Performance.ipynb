### 1) Problem statement
- This project understands how the student's performance (test scores) is affected by other variables such as Gender, Ethnicity, Parental level of education, Lunch and Test preparation course.


### 2) Data Collection
- Dataset Source - https://www.kaggle.com/datasets/spscientist/students-performance-in-exams?datasetId=74977
- The data consists of 8 column and 1000 rows.

### 2.1 Import Data and Required Packages

- lunch : having lunch before test (standard or free/reduced) 
- test preparation course : complete or not complete before test
- math score

- writing score

```python
# Adding columns for "Total Score" and "Average"

df['total score'] = df['math_score'] + df['reading_score'] + df['writing_score']
df['average'] = df['total score']/3
df.head()
```

```python
reading_full = df[df['reading_score'] == 100]['average'].count()
writing_full = df[df['writing_score'] == 100]['average'].count()
math_full = df[df['math_score'] == 100]['average'].count()

print(f'Number of students with full marks in Maths: {math_full}')
print(f'Number of students with full marks in Writing: {writing_full}')
print(f'Number of students with full marks in Reading: {reading_full}')
```

```python
reading_less_20 = df[df['reading_score'] <= 20]['average'].count()
writing_less_20 = df[df['writing_score'] <= 20]['average'].count()
math_less_20 = df[df['math_score'] <= 20]['average'].count()

print(f'Number of students with less than 20 marks in Maths: {math_less_20}')
print(f'Number of students with less than 20 marks in Writing: {writing_less_20}')
print(f'Number of students with less than 20 marks in Reading: {reading_less_20}')
```

#####  Insights
 - From above values we get students have performed the worst in Maths 
 - Best performance is in reading section

##### Maximumum score of students in all three subjects

```python
# Set the overall size of the entire figure (canvas)
# figsize=(14,6) means:
# width = 14 inches
# height = 6 inches
plt.figure(figsize=(14,6))


# Create first subplot
# (1, 4, 1) means:
# 1 row
# 4 columns
# This is the 1st plot position
plt.subplot(1, 4, 1)
# Set title for the first subplot
plt.title('MATH SCORES')
# Draw a violin plot
# y='math_score' → column name from dataframe
# data=df → use dataframe df
# color='red' → fill color of violin
# linewidth=3 → thickness of boundary line
sns.violinplot(y='math_score', data=df, color='red', linewidth=3)


# Create second subplot
# Still 1 row and 4 columns
# This is the 2nd plot position
plt.subplot(1, 4, 2)
# Set title
plt.title('READING SCORES')
# Violin plot for reading scores
sns.violinplot(y='reading_score', data=df, color='green', linewidth=3)


# Create third subplot
# This is the 3rd position in same row
plt.subplot(1, 4, 3)
# Set title
plt.title('WRITING SCORES')
# Violin plot for writing scores
sns.violinplot(y='writing_score', data=df, color='blue', linewidth=3)


# Display all plots together
plt.show()
```

#### Multivariate analysis using pieplot

```python
# Set default figure size for all plots
# (30,12) → width=30 inches, height=12 inches
plt.rcParams['figure.figsize'] = (30, 12)


# -------------------- 1 Gender Pie Chart --------------------

# Create subplot grid:
# 1 row, 5 columns, position 1
plt.subplot(1, 5, 1)

# Count number of males and females
size = df['gender'].value_counts()

# Labels for pie slices
labels = 'Female', 'Male'

# Colors for slices
color = ['red','green']

# Draw pie chart
# size → values
# colors → slice colors
# labels → slice names
# autopct → display percentage
plt.pie(size, colors=color, labels=labels, autopct='.%2f%%')

# Title of subplot
plt.title('Gender', fontsize=20)

# Remove axis lines
plt.axis('off')



# -------------------- 2 Race/Ethnicity Pie Chart --------------------

plt.subplot(1, 5, 2)

# Count each race category
size = df['race_ethnicity'].value_counts()

# Labels for groups
labels = 'Group C', 'Group D','Group B','Group E','Group A'

# Colors for slices
color = ['red', 'green', 'blue', 'cyan','orange']

# Draw pie chart
plt.pie(size, colors=color, labels=labels, autopct='.%2f%%')

plt.title('Race/Ethnicity', fontsize=20)

plt.axis('off')



# -------------------- 3 Lunch Type Pie Chart --------------------

plt.subplot(1, 5, 3)

# Count lunch types
size = df['lunch'].value_counts()

labels = 'Standard', 'Free'

color = ['red','green']

plt.pie(size, colors=color, labels=labels, autopct='.%2f%%')

plt.title('Lunch', fontsize=20)

plt.axis('off')



# -------------------- 4️ Test Preparation Course --------------------

plt.subplot(1, 5, 4)

# Count preparation course categories
size = df['test_preparation_course'].value_counts()

labels = 'None', 'Completed'

color = ['red','green']

plt.pie(size, colors=color, labels=labels, autopct='.%2f%%')

plt.title('Test Course', fontsize=20)

plt.axis('off')



# -------------------- 5 Parental Education Pie Chart --------------------

plt.subplot(1, 5, 5)

# Count each education level
size = df['parental_level_of_education'].value_counts()

labels = 'Some College', "Associate's Degree",'High School',
         'Some High School',"Bachelor's Degree","Master's Degree"

color = ['red', 'green', 'blue', 'cyan','orange','grey']

plt.pie(size, colors=color, labels=labels, autopct='.%2f%%')

plt.title('Parental Education', fontsize=20)

plt.axis('off')



# Adjust spacing between plots automatically
plt.tight_layout()

# Add grid (not very useful for pie charts)
plt.grid()

# Display all plots
plt.show()
```

#####  Insights
- Number of Male and Female students is almost equal
- Number students are greatest in Group C
- Number of students who have standard lunch are greater
- Number of students who have not enrolled in any test preparation course is greater
- Number of students whose parental education is "Some College" is greater followed closely by "Associate's Degree"

#### 4.4 Feature Wise Visualization
#### 4.4.1 GENDER COLUMN
- How is distribution of Gender ?
- Is gender has any impact on student's performance ?

#### UNIVARIATE ANALYSIS ( How is distribution of Gender ? )

```python
# Create figure with 1 row and 2 columns
# f → figure object
# ax → array of axes (2 plots)
# figsize=(8,4) → width=8 inches, height=4 inches
f, ax = plt.subplots(1, 2, figsize=(8, 4))


# -------------------- 1 Count Plot --------------------

# sns.countplot()
# x=df['gender'] → categorical column
# data=df → dataframe
# palette='bright' → color theme
# ax=ax[0] → draw plot in first subplot
# saturation=0.95 → color intensity
sns.countplot(x=df['gender'], data=df,
              palette='bright',
              ax=ax[0],
              saturation=0.95)


# Add number labels on top of bars
# ax[0].containers → all bars
for container in ax[0].containers:
    ax[0].bar_label(container, color='black', size=20)


# -------------------- 2 Pie Chart --------------------

# Create pie chart
# x=df['gender'].value_counts() → count of males & females
# labels → category names
# explode=[0,0.1] → slightly separate second slice
# autopct='%1.1f%%' → show percentage with 1 decimal
# shadow=True → adds shadow effect
# colors → custom slice colors
plt.pie(x=df['gender'].value_counts(),
        labels=['Male','Female'],
        explode=[0,0.1],
        autopct='%1.1f%%',
        shadow=True,
        colors=['#ff4d4d','#ff8000'])


# Display the plots
plt.show()
```

#### Insights 
- Gender has balanced data with female students are 518 (48%) and male students are 482 (52%) 

#### BIVARIATE ANALYSIS ( Is gender has any impact on student's performance ? ) 

```python
gender_group = df.groupby('gender').mean(numeric_only=True)
gender_group
```

```python
plt.figure(figsize=(7, 4))

X = ['Total Average','Math Average']


female_scores = [gender_group['average'][0], gender_group['math_score'][0]]
male_scores = [gender_group['average'][1], gender_group['math_score'][1]]

X_axis = np.arange(len(X))
  
plt.bar(X_axis - 0.2, male_scores, 0.4, label = 'Male')
plt.bar(X_axis + 0.2, female_scores, 0.4, label = 'Female')
  
plt.xticks(X_axis, X)
plt.ylabel("Marks")
plt.title("Total average v/s Math average marks of both the genders", fontweight='bold')
plt.legend()
plt.show()
```

#### Insights 
- On an average females have a better overall score than men.
- whereas males have scored higher in Maths.
